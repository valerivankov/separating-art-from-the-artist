# The Effects of Knowledge About an Artist on Art Preferences

## Introduction

Art, a form of creative skill and expression, has a ubiquitous impact on culture, politics, and economics (Schleter 2011). Since the rise of human civilization, art has shaped both the individual and collective thought and has significantly contributed to the functioning of contemporary society (Talley 2023). As a result, understanding how art is perceived can provide insight into the global effects of this phenomenon and its relation to a variety of social movements. The following project will focus on discerning the complex connection between art and the artist. The main question this paper poses is whether humans can separate art from the artist. More specifically, the primary goal is to determine the extent to which one's attitude towards an artist can affect their preference toward an artwork. The individual's ability to isolate an artwork from its creator has sparked debate amongst humanists and social scientists; at the same time, the actions of controversial artists, like Gaugin and Picasso, have often created division among supporters of their works (Kim 2022). Despite the abundance of philosophical discourse on the issue, little research has been done to formally support either side of the dilemma. Even the few attempts of answering the question in a controlled environment have been limited by a number of factors, ranging from the exclusion of people with artistic backgrounds to the comparison of negative and neutral pieces of knowledge only to demonstrate the causal effect of opinion on an artist on the liking of an art piece. (Kaube, Eiserbeck, and Abdel Rahman 2023). Hence, the next section will focus on presenting a novel approach to the dilemma whilst accounting for a number of different factors.

## Experimental Design

As stated above, the study will use a randomized block design with respect to artists (people with formal background in art or practitioners of the craft) and non-artists. It is important to differentiate between the two groups since artists tend to have more experience analyzing images than nonartists. Additionally, the preferences of artists are likely to be influenced by artistic qualities of the artwork that cannot be recognized by nonartists. In contrast, a nonartist will more often than not judge an art piece through a subjective lens rather than objective criteria (Hawley-Dolan and Winner 2011; Vogt and Magnussen 2007).

Furthermore, the study will control for the nature of the statements that are given to the participants. This will happen in two directions:
- Connotation of the statement: **positive**, **neutral**, or **negative**
- Relatedness of the statement: **related** or **unrelated**

To ensure that positive and negative statements along with related and unreleated statements have the same strength, a separate prior study will be conducted. Moreover, all of the statements will be given in the form of one-sentence declaritive claims about an artist that will have a similar number of words.

Lastly, it should be noted that, unlike previous research, which utilizes real paintings and thus creates the risk of subjecting people, especially artists, to artworks that they are already familiar with, this experiment will deal with generative art (that is, art created in part or entirely by artificial intelligence). Generative art has multiple advantages: it allows researchers to create a narrative around an artwork; it is indistinguishable from art created by humans; it can be customized to follow a specific style and theme (Benney and Kistler 2021; Elgammal et al. 2017).

The project will now outline the experimental procedure while the following sections will simulate and analyze fake participant data, accordingly, in order to predict the actual experimental results.

The study will recruit 72 artists and 72 non-artists (sample size is explained in more detail in the "Power Analysis" section of the paper). In total 144 paintings generated by an artificial intelligence software will be used. The participants will be randomly divided in two equally sized groups of 72, so that each group has 36 artists and 36 non-artists. Each participant in the first (control) group will be presented the images without any supporting information while each participant in the second (treatment) group will be presented the images with the corresponding statement about the image. To avoid bias regarding the order in which the paintings were shown to the participants, the experiment will employ a Latin square design, meaning that if the paintings are denoted P1, P2, ..., P144, participant 1 will be shown paintings P1, P2, ..., P144 in this order, participant 2 will be shown paintings P2, P3, ..., P144, P1 in this order, etc. Such a design would require randomizing the order of paintings for the first participant, so that the order is uniquely determined for the rest of the participants and is still random. The experiment is thus counterbalanced in terms of the fact that each participant will see all images and the order of seeing them will be different for each of them (Pawe≈Ç Ciborowski and Jerzy Silberring 2016). Images will be presented for a short period of time (three seconds) and participants will be asked to evaluate them based on their initial reaction. The participants will rate the paintings on a scale from 1 to 10 (1 being the lowest in terms of liking and 10 being the highest). The difference of means between groups will be calculated to estimate the effect of providing knowledge about an artist on a person's perception of a painting. 

The 144 paintings were selected on the basis of several rules. Images with ambiguous messaging (including abstract paintings) were excluded in order to avoid the risk of subjective interpretation of the relatedness of the statements and the image they were paired with. In addition, the paintings were divided in 24 sets of 6 paintings in terms of matching theme, style, and complexity so as to control for potential differences in top-down and bottom-up processing of sensory information (Delorme et al. 2004; Sobel et al. 2007). All paintings in the same group display similar artistic skill or approach as well as a common theme, such as "forests." However, each painting within the set is associated with a different combination of connotation and relatedness pairing so that all six possible combinations (positve-related, neutral-related, negative-related, positive-unrelated, neutral-unrelated, negative-unrelated) are represented. Below is an example of what the six statements for a set of paintings might be based on the theme of "forests":

|                  | **Positive**                     | **Neutral**                   | **Negative**                 |
| ---------------- | ------------------------------ | ---------------------------- | ---------------------------- |
| **Related**      | The author plants trees.       | The author lives near a forest. | The author supports deforestation. |
| **Unrelated**    | The author is an animal rights activist. | The author plays the piano. | The author has misogynistic views. |

The similarity of themes, style, and complexity of paintings will be assessed prior to the actual experiment.

Given all of the above, the paper has three main hypotheses:
- The connotation of the statement will have a proportional effect on the average rating of a painting. In other words, positive statements will, on average, increase the rating of a painting while negative statements will decrease it. Neutral statements are not expected to affect the rating of an image.
- Unrelatedness will not affect the rating of an image or interfere with the effect of the connotation of the statement. Relatedness, on the other hand, will enhance the effect of the connotation of the statement, provided that it is positive or negative. In other words, positive related and negative related statements will, on average, have a stronger effect than positive unrelated and negative unrelated statements, respectively.
- As a result of their prior experience, artists are expected to be less affected by the nature of the statement that is presented to them than non-artists. Thus, a positive or negative statement, regardless of relatedness, will have a weaker effect on the ratings of artists than the ratings of non-artists.

The project proceeds with a simulation of numerous fake trials. The average treatment effect of each trial will be calculated and saved for further analysis.

## Simulation of Data

We will create a new data frame with all the relevant fields:

```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
import statsmodels.stats.power as smp

num_rows = 144*144

sim_data = pd.DataFrame({
    'participant': [None] * num_rows,
    'outcome': [None] * num_rows,
    'treatment_type': [None] * num_rows,
    'connotation': [None] * num_rows,
    'relatedness': [None] * num_rows,
    'block': [None] * num_rows,
    'connotation_effect': [None] * num_rows,
    'relatedness_effect': [None] * num_rows,
    'block_effect': [None] * num_rows,
    'error': [None] * num_rows
})
```

We start by filling in some of the columns:

```python
# Participant Number
for i in range(1, 145):
    start_index = (i - 1) * 144
    end_index = i * 144
    sim_data.loc[start_index:end_index, 'participant'] = i

# Treatment Type (0 for control, 1 for treatment)
sim_data['treatment_type'] = np.where(sim_data['participant'] <= 72, 1, 0)

# Block (Artist or Non-Artist)
sim_data['block'] = np.where(sim_data['participant'] % 2 == 0, 'artist', 'non-artist')

# Error Term
sim_data['error'] = np.concatenate([np.random.normal(0, 0.001, 72*144), np.zeros(72*144)])
```

Since there are 24 sets of 6 paintings, each containing all possible pairings of connotation and relatedness, this means that we have, in total, 24 paintings representing each of the possible combinations. Hence, there are 48 statements by connotation and 72 statements by relatedness.

Given this information, we will now fill in the "connotation" and "relatedness" columns in a random order for the first participant:

```python
connotation_choices = np.concatenate([np.repeat("positive", 48), np.repeat("neutral", 48), np.repeat("negative", 48)])
relatedness_choices = np.concatenate([np.repeat("related", 72), np.repeat("unrelated", 72)])

np.random.shuffle(connotation_choices)
np.random.shuffle(relatedness_choices)

participant_1_indices = sim_data['participant'] == 1
sim_data.loc[participant_1_indices, 'connotation'] = connotation_choices[:participant_1_indices.sum()]
sim_data.loc[participant_1_indices, 'relatedness'] = relatedness_choices[:participant_1_indices.sum()]
```

We save the order of the information combinations for the first participant and modify it for the rest of the 71 participants that get the treatment according to the Latin square design principle.

```python
con_1 = sim_data.loc[sim_data['participant'] == 1, 'connotation'].values
rel_1 = sim_data.loc[sim_data['participant'] == 1, 'relatedness'].values

for i in range(2, 145):
    participant_i_indices = sim_data['participant'] == i
    sim_data.loc[participant_i_indices, 'connotation'] = np.concatenate((con_1[i:145], con_1[0:i]))
    sim_data.loc[participant_i_indices, 'relatedness'] = np.concatenate((rel_1[i:145], rel_1[0:i]))
```

We can now fill in the "connotation_effect," "relatedness_effect," and "block_effect" columns. Based on similar studies, we can estimate the average effective of a positive or negative effect to be approximately 0.36 or -0.36, respectively. The relatedness effect is 0.1 if the statement is positive and -0.1 if the statement is negative. There is no interaction between connotation and relatedness. Finally, since the experiment is blocked by art background, we will add a "neutralizing" coefficient of 0.05; that is, if the statement was positive, the coefficient will equal -0.05 for artists; if the statement was negative, the coefficient will equal 0.05; the coefficient will be left at 0 otherwise. Let us implement these effects in the data frame:

```python
for i in range(1, 73):
    sim_data.loc[(sim_data['participant'] == i) & (sim_data['connotation'] == "positive"), 'connotation_effect'] = 0.36
    sim_data.loc[(sim_data['participant'] == i) & (sim_data['connotation'] == "negative"), 'connotation_effect'] = -0.36
    sim_data.loc[(sim_data['participant'] == i) & (sim_data['connotation'] == "neutral"), 'connotation_effect'] = 0

for i in range(1, 73):
    sim_data.loc[(sim_data['participant'] == i) & (sim_data['relatedness'] == "related"), 'relatedness_effect'] = 0.1 * np.sign(sim_data.loc[sim_data['participant'] == i, 'connotation_effect'])
    sim_data.loc[(sim_data['participant'] == i) & (sim_data['relatedness'] == "unrelated"), 'relatedness_effect'] = 0

for i in range(2, 73, 2):
    sim_data.loc[(sim_data['participant'] == i), 'block_effect'] = -0.05 * np.sign(sim_data.loc[(sim_data['participant'] == i), 'connotation_effect'].values[0])

for i in range(1, 72, 2):
    sim_data.loc[(sim_data['participant'] == i), 'block_effect'] = 0

for i in range(73, 145):
    sim_data.loc[(sim_data['participant'] == i), ['connotation_effect', 'relatedness_effect', 'block_effect']] = 0
```

Finally, we can calculate the outcome. For each participant and painting combination, we will randomly select a rating from 1 to 10 and add all of the effects:

```python
# Creating ratings vector
ratings = np.random.randint(1, 11, size=144*144)

# Calculating outcome variable
sim_data['outcome'] = ratings + sim_data['treatment_type'] * (sim_data['connotation_effect'] + sim_data['relatedness_effect'] + sim_data['block_effect']) + sim_data['error']
```

## Data Analysis

We can evaluate the average treatment effect of ratings by doing 1000 trials and storing the results in a vector called overall_rating_effects. Likewise, we will obtain similar vectors for positive statements (positive_effects) and negative statements (negative_effects) by finding the average difference between the outcomes of paintings associated with positive (or negative) statements and the average rating in the control group:

```python
N = 1000

overall_rating_effects = np.empty(N)
positive_effects = np.empty(N)
negative_effects = np.empty(N)

for i in range(N):
    ratings = np.random.randint(1, 11, size = 144*144)
    
    sim_data['outcome'] = (
        ratings +
        sim_data['treatment_type'] * (
            sim_data['connotation_effect'] +
            sim_data['relatedness_effect'] +
            sim_data['block_effect']) +
        sim_data['error'])
    
    overall_rating_effects[i] = (
        np.mean(sim_data[sim_data['treatment_type'] == 1]['outcome']) -
        np.mean(sim_data[sim_data['treatment_type'] == 0]['outcome']))
    
    positive_effects[i] = (
        np.mean(sim_data[(sim_data['treatment_type'] == 1) & (sim_data['connotation'] == 'positive')]['outcome']) -
        np.mean(sim_data[sim_data['treatment_type'] == 0]['outcome']))
    
    negative_effects[i] = (
        np.mean(sim_data[(sim_data['treatment_type'] == 1) & (sim_data['connotation'] == 'negative')]['outcome']) -
        np.mean(sim_data[sim_data['treatment_type'] == 0]['outcome']))
```

The means of the estimates are

```python
print(np.mean(overall_rating_effects))
print(np.mean(positive_effects))
print(np.mean(negative_effects))
```

Unsurprisingly, the mean of overall effect is close to zero while the means of the ratings of paintings with positive and negative information pairings are positive and negative, respectively.

The standard error of the estimate is

```python
print(np.std(overall_rating_effects))
print(np.std(positive_effects))
print(np.std(negative_effects))
```

The standard deviations of the change in ratings of paintings presented with positive and negative statement are very similar. Moreover, the standard deviation of the average treatment effect of the entire experiment is slightly lower.

We calculate the p-value using a two-sided Welch test against the null hypothesis that the average treatment effect is zero. The test is suitable since, by the Central Limit Theorem, the distribution of the means of all distributions will be approximately normal.

```python
t_stat_overall, p_value_overall = stats.ttest_1samp(overall_rating_effects, 0, alternative = 'two-sided')
t_stat_positive, p_value_positive = stats.ttest_1samp(positive_effects, 0, alternative = 'two-sided')
t_stat_negative, p_value_negative = stats.ttest_1samp(negative_effects, 0, alternative = 'two-sided')

print(p_value_overall)
print(p_value_positive)
print(p_value_negative)
```

As expected, the p-value is above 0.05 for overall effects and less than 0.05 for the other two cases, meaning that overall there is no effect of the nature of the knowledge about an artist on preferences about art. This is because, no matter the background of the participant or the relatedness of the statements, positive statements and negative statements have negating effects. If we filter by a specific types of positive or negative knowledge about an artist, however, we get significant results about the inability to separate the art from the artist.


## Power Analysis

The study aimed to achieve a power of at least 0.9. Since the main effects of interests are those of positive and negative statements, the chosen effect size is 0.4. At a significance level 0.05, the minimal required sample size can be calculated by the following function:

```python
effect_size = 0.4
alpha = 0.05
power = 0.9

print(smp.TTestIndPower().solve_power(effect_size = effect_size, alpha = alpha, power = power, alternative = 'two-sided'))
```

Thus N = 144 fulfills the power analysis criterion.

## Conclusion

Based on the pre-analysis of this study, it appears that people tend to be influenced by their knowledge of an artist whenever they assess an artwork. However, it should be acknowledged that the reality of the psychological processes that occur when engaging with art are far more nuanced. For one thing, the statements were only judged on the basis of connotation and relatedness, whereas in practice other factors might have a stronger impact on one's evaluation of an artwork given the context of the artists' thoughts and actions. Moreover, while some controversial artists have been "called out" or "cancelled" for their wrongdoing, especially during the age of social media, others have still remained popular because even of people do not agree with an artist's actions, they are still capable of recognizing their talent and skill (France-Presse 2022). Thus more research needs to be conducted in order to truly unravel the complex relationship between the artist, the artworks, and the audience.
